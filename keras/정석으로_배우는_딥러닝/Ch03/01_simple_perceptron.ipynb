```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 난수의 상태 결정. 
rng = np.random.RandomState(123)

# d : input의 갯수
# N : data의 갯수
# mean : Threshold
d = 2
N = 10
mean = 5

# rng.randn(N, d) : seed 값으로 난수를 뽑아서 [10, 2] 행렬 생성.
x1 = rng.randn(N, d) + np.array([0,0])
x2 = rng.randn(N, d) + np.array([5,5])
```

```python
x = np.concatenate([x1, x2], axis=0)
```

```python
plt.plot(x1[:,0], x1[:,1], 'rx')
plt.plot(x2[:,0], x2[:,1], 'bo')
plt.legend(['x1','x2'],loc=0)
plt.show()
```

```python
# Initialization W, b
w = np.zeros(d)
b = 0
```

```python
def step(x):
    '''
    Step Function
    '''
    return 1*(x>0)

def y(x):
    '''
    y = Step(w * x + b)
    '''
    
    return step(np.dot(w, x)+b)
```

```python
def t(i):
    '''
    Labeling
    i<N -> x[:10] = 0
    i>=N -> x[10:] = 1
    '''
    if i<N:
        return 0
    else:
        return 1
```

```python
n = 0
while True:
    classified = True
    for i in range(N*2):
        
        cost = (t(i) - y(x[i]))
        delta_w = cost * x[i]
        delta_b = cost * 1
        
        w += delta_w
        b += delta_b
        
        classified *= all(delta_w == 0) * (delta_b == 0)
        
        print('Epochs : ', n+1)
        print('Step   : ', i+1)
        print("Diff_w : ", delta_w)
        print("Diff_b : ", delta_b)
        print('Updated w : ', w)
        print('Updated b : ', b)
        print('Classified : ', classified)
        print('------------------------------\n')
        
        
        
    if classified:
        break
    n+=1
```
